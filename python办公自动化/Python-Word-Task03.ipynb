## 初步认识docx

```python
# 导入库
from docx import Document

# 新建空白文档，声明对象
doc_1 = Document()

# 添加标题 (0 相当于文章的题目， 默认级别是1，  级别范围是0-9)
doc_1.add_heading('新建空白文档标题，级别为0', level = 0)
doc_1.add_heading('新建空白文档标题，级别为1', level = 1)
doc_1.add_heading('新建空白文档标题，级别为2', level = 2)

# 新增段落
paragraph_1 = doc_1.add_paragraph('这个是我的第一个python-word项目，哈哈哈哈\n')

# 加粗, 这里是直接设置它的属性
paragraph_1.add_run('加粗字体').bold = True
paragraph_1.add_run('普通字体')

# 斜体
paragraph_1.add_run('斜体').italic = True

# 新加入一个段落
paragraph_2 = doc_1.add_paragraph('第二个段落，也不知道对不对')

# 指定段落插入位置
prior_paragraph = paragraph_1.insert_paragraph_before('在第一段文字前插入段落')
#after_paragraph2 = paragraph_2.insert_paragraph_after('在原来第二段文字后插入段落')

# 添加分页符
doc_1.add_page_break()

# 新段落
paragraph_3 = doc_1.add_paragraph('这个是第二页的第一段文字')

# 保存文字
doc_1.save('doc_1.docx')
```

## docx 在python 的页面结构

```python
# 了解Word 文档的页面结构
# document ——> paragraph ——> run

# 导入库
from docx import Document
from docx.shared import RGBColor, Pt, Inches, Cm
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.oxml.ns import qn

# 新建文档

doc_2 = Document()


# 字体设置（全局）
# 这里不只是设置font.name, 还需要调用.element.rPr.rFonts的set()方法

doc_2.styles['Normal'].font.name = u'宋体'
doc_2.styles['Normal']._element.rPr.rFonts.set(qn('w:eastAsia'), u'宋体')

# 添加标题 
heading_1 = doc_2.add_heading("这个是第二次创建标题哦", level = 0)
heading_1.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER  # 居中对齐， 默认居左对齐


# 新增段落
paragraph_1 = doc_2.add_paragraph()

# 设置段落的格式
'''
设置段落格式：首行缩进0.75cm，居左，段后距离1.0英寸,1.5倍行距。
'''
paragraph_1.paragraph_format.first_line_indent = Cm(0.75)
paragraph_1.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT
paragraph_1.paragraph_format.space_after = Inches(1.0)
paragraph_1.paragraph_format.line_spacing = 1.5


text = '中国台湾华语流行歌手、' \
'音乐创作家、作曲家、作词人、' \
'制作人、杰威尔音乐公司老板之一、导演。' \
'近年涉足电影行业。周杰伦是2000年后亚洲流行乐坛最具革命性与指标' \
'性的创作歌手，有“亚洲流行天王”之称。他突破原有亚洲音乐的主题、形' \
'式，融合多元的音乐素材，创造出多变的歌曲风格，尤以融合中西式曲风的嘻哈' \
'或节奏蓝调最为著名，可说是开创华语流行音乐“中国风”的先声。周杰伦的' \
'出现打破了亚洲流行乐坛长年停滞不前的局面，为亚洲流行乐坛翻开了新的一页'

r_l = paragraph_1.add_run(text)
r_l.font.size = Pt(10)
r_l.font.bold = True
r_l.font.color.rgb = RGBColor(25, 23, 88)

# 保存文件
doc_2.save('doc_2.docx')
```

```python
# 字体设置

# 导入库
from docx import Document
from docx.oxml.ns import qn
from docx.enum.style import WD_STYLE_TYPE

doc_3 = Document()

# 设置宋体字样
style_font = doc_3.styles.add_style('宋体', WD_STYLE_TYPE.CHARACTER)
style_font.font.name = '宋体'
doc_3.styles['宋体']._element.rPr.rFonts.set(qn('w:eastAsia'), '宋体')

# 设置楷体字样
style_font = doc_3.styles.add_style('楷体', WD_STYLE_TYPE.CHARACTER)
style_font.font.name = '楷体'
doc_3.styles['楷体']._element.rPr.rFonts.set(qn('w:eastAsia'), '楷体')


# 设置为 华文中宋字样
style_font = doc_3.styles.add_style('华文中宋', WD_STYLE_TYPE.CHARACTER)
style_font.font.name = '华文中宋'
doc_3.styles['华文中宋']._element.rPr.rFonts.set(qn('w:eastAsia'), '华文中宋')


paragraph1 = doc_3.add_paragraph()  # 添加段落
run = paragraph1.add_run(u'abCD 这是我的python-excel', style = '宋体')

font = run.font
font.name = 'Cabira'
paragraph1.add_run(u'abCD 这是我的python-excel', style = '楷体').font.name
paragraph1.add_run(u'abCD 这是我的python-excel', style = '华文中宋').font.name

doc_3.save('font-setting.docx')
```

```python
import openpyxl

wb = openpyxl.load_workbook('excelToword.xlsx')
sheet = wb.active
cells = sheet['A1:D5']
i = 0
for row in sheet.iter_rows(min_row = 2, max_row = 5,
                           min_col = 1, max_col =4):
    i = i + 1
    produce_doc(i, row[0].value, row[1].value, row[2].value, row[3].value)
```

```python
# 导入库
from docx import Document
from docx.shared import RGBColor, Pt, Inches, Cm
from docx.enum.text import WD_PARAGRAPH_ALIGNMENT
from docx.oxml.ns import qn
from docx.shared import RGBColor,Pt

def produce_doc(i, r0, r1, r2, r3):

    doc = Document()

    heading_1 = doc.add_heading("邀请函", level = 0)
    heading_1.alignment = WD_PARAGRAPH_ALIGNMENT.CENTER

    # 新增段落
    paragraph_1 = doc.add_paragraph()

    # 设置段落的格式
    '''
    设置段落格式：首行缩进0.75cm，居左，段后距离1.0英寸,1.5倍行距。
    '''
    paragraph_1.paragraph_format.first_line_indent = Cm(0.75)
    paragraph_1.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.LEFT
    paragraph_1.paragraph_format.space_after = Inches(1.0)
    paragraph_1.paragraph_format.line_spacing = 1.5
    paragraph_1.add_run("尊敬%s的公司%s%s，您好" %(r0, r1, r2))
    paragraph_1.add_run('just for testing').font.underline = True

    paragraph_2 = doc.add_paragraph()
    paragraph_2.add_run('邀请您参加活动！')


    paragraph_3 = doc.add_paragraph()
    paragraph_3.paragraph_format.alignment = WD_PARAGRAPH_ALIGNMENT.RIGHT
    paragraph_3.add_run('邀请时间：%s' %r3)
    doc.save('test%s.docx' %i)
```
